### Q.2: Write a function `myFP` that takes in the fixed point iteration function handle `g`, the initial point `x0` and performs the fixed point iterations $x_{n+1} = g(x_n)$ till the error estimate: $r - x_n = \frac{g'(c_n)}{1-g'(c_n)} (x_n - x_{n-1})$, goes below a set threshold. You can use the estimate $g'(c_n) \simeq \frac{x_n -x_{n-1}}{x_{n-1}-x_{n-2}}$ in order to estimate $r-x_n$. The function should return the sequence $\{x_n\}_{n=0}^{N}$.

```python
# def myFP(g,x0):


def myFP(g, x0):
    max_iterations = 100 # setting the maximum number of iterations as 100 in case of unbounded iterations 
    threshold = 1e-6 # the set threshold
    xn = [x0]  # array for the fixed points, first element would be the assumed x0
    x = [0]  # Counter for the number of iterations
    for i in range(max_iterations):
        x.append(i+1)  # iteration number increment
        xn.append(g(xn[-1]))  # appending to the end of the array
        if i >= 2:
            # iterating from the back of the array
            g_dash = (xn[-1] - xn[-2])/(xn[-2] - xn[-3])
            if (1 - g_dash != 0):
                error_estimate = abs(g_dash/(1 - g_dash)*(xn[-1] - xn[-2]))
            else:
                # used to remove Zero Divison error in case 1-g_dash=0, abs used to get the positive value of the difference
                error_estimate = abs(xn[-1] - xn[-2])

            if error_estimate < threshold:
                break  # if the error estimate goes below the set threshold  =1e-6

    plt.plot(x, xn, marker='o', label='fixed points')
    plt.xlabel('Fixed point estimates')
    plt.ylabel('Iteration number')
    plt.title('Xn v/s n graph')
    return xn
```

```python
def g(x):
    return 1/2*np.exp(-x)
print(myFP(g,1))
```
